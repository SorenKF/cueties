```python
'''creating a dummy input of a list of tuples of tuples of indices. Where the the list represents the dataset, the first
layer of tuple represents the paired instances and the second layer of tuples represent the start and end indices of the
source and content spans'''

list_of_tuples = list()

source1 = (0,1 )
content1 = (7,14)
source2 = (50,53)
content2 = (55, 65)
source3 = (250, 252)
content3 = (255, 265)
source4 = (1340, 1342)
content4 = (1346, 1352)


tuple1 = (source1, content1)
tuple2 = (source2, content2)
tuple3 = (source3, content3)
tuple4 = (source4, content4)

list_of_tuples.append(tuple1)
list_of_tuples.append(tuple2)
list_of_tuples.append(tuple3)
list_of_tuples.append(tuple4)

print(list_of_tuples)
```

```python
'''features to attempt are: 
number of words between c and s / 
number of contents between c and s /
number of entity mentions between c and s'''
```

```python
def distance_in_tokens(list_of_tuples, df):
    '''function to find the distance in tokens between the content and source
    takes - the list of tuples of tuples of indices as a variable.
          - the df being used as a variable
    returns - the df with an added column s/c_distance in which all tokens in the source/content span have the distance number
    and all other tokens have X
    '''
    df['s/c_distance'] = 'X'
    
    for instance in list_of_tuples:
        instance_list = list()
        index_list = list()
        
        source = instance[0]
        b1 = source[0]
        e1 = source[1]
        index_list.append(e1)
        for item in range(b1, e1):
            index_list.append(item)
        
        if e1 > b1:
            instance_list.append(e1)
        else:
            instance_list.append(b1)
        
        content = instance[1]
        b2 = content[0]
        e2 = content[1]
        index_list.append(e2)
        for item in range(b2, e2):
            index_list.append(item)
        
        for index in instance_list:
            if index < b2:
                if b2 < e2:
                    instance_list.append(b2)
                else:
                    instance_list.append(e2)
            elif index > b2:
                if b2 < e2:
                    instance_list.append(e2)
                else:
                    instance_list.append(b2)
        
        if instance_list[0] < instance_list[1]:
            distance = instance_list[1] - instance_list[0]
        else:
            distance = instance_list[0] - instance_list[1]
   
        df.loc[index_list,'s/c_distance'] = distance
```

```python
def num_contents_between(list_of_tuples, df):
    '''function to find the number of contents between the source and content pair in the instance
    takes - the list of tuples of tuples of indices as a variable
          - the df being used as a variable
    returns - the df with a new column added num_conts_between in which all tokens in the source/content span have the 
    number of contents between the two and all other tokens have X'''
    
    df['num_conts_between'] = 'X'
    
    for instance in list_of_tuples:
        instance_list = list()
        index_list = list()
        
        source = instance[0]
        b1 = source[0]
        e1 = source[1]
        index_list.append(e1)
        for item in range(b1, e1):
            index_list.append(item)
        
        if e1 > b1:
            instance_list.append(e1)
        else:
            instance_list.append(b1)
        
        content = instance[1]
        b2 = content[0]
        e2 = content[1]
        index_list.append(e2)
        for item in range(b2, e2):
            index_list.append(item)
        
        for index in instance_list:
            if index < b2:
                if b2 < e2:
                    instance_list.append(b2)
                else:
                    instance_list.append(e2)
            elif index > b2:
                if b2 < e2:
                    instance_list.append(e2)
                else:
                    instance_list.append(b2)
                    
        #print(instance_list)
        counter = 0
        for index in list(range(instance_list[0], instance_list[1])):
            if df.loc[index,'content_label_gold'] == 'B':
                counter += 1
                
        df.loc[index_list,'num_conts_between'] = counter
```

```python
def create_content_source_dist_feats(list_of_tuples, df):
    '''A function to add all content/source distance features to the df
    takes- list_of_tuples = a list of tuples of tuples of indices of the paired source/content spans
         - df = the dataframe containing the dataset
    returns- the df with added s/c_distance, num_conts_between and num_sources_between columns. Where X is used for tokens
    outside of the s/c pairs and numbers are given to all tokens inside of the s/c span to show the information in column'''
    
    df['s/c_distance'] = 'X'
    df['num_conts_between'] = 'X'
    df['num_sources_between'] = 'X'
    
    for instance in list_of_tuples:
        instance_list = list()
        index_list = list()
        
        source = instance[0]
        b1 = source[0]
        e1 = source[1]
        index_list.append(e1)
        for item in range(b1, e1):
            index_list.append(item)
        
        if e1 > b1:
            instance_list.append(e1)
        else:
            instance_list.append(b1)
        
        content = instance[1]
        b2 = content[0]
        e2 = content[1]
        index_list.append(e2)
        for item in range(b2, e2):
            index_list.append(item)
        
        for index in instance_list:
            if index < b2:
                if b2 < e2:
                    instance_list.append(b2)
                else:
                    instance_list.append(e2)
            elif index > b2:
                if b2 < e2:
                    instance_list.append(e2)
                else:
                    instance_list.append(b2)
        
        if instance_list[0] < instance_list[1]:
            distance = instance_list[1] - instance_list[0]
        else:
            distance = instance_list[0] - instance_list[1]
   
        df.loc[index_list,'s/c_distance'] = distance
    
        counter = 0
        for index in list(range(instance_list[0], instance_list[1])):
            if df.loc[index,'content_label_gold'] == 'B':
                counter += 1
                
        df.loc[index_list,'num_conts_between'] = counter
        
        counter = 0
        for index in list(range(instance_list[0], instance_list[1])):
            if df.loc[index,'candidate_source'] == 'B':
                counter += 1
                
        df.loc[index_list,'num_sources_between'] = counter
        
    return df
```

```python
create_content_source_dist_feats(list_of_tuples, df)
```
