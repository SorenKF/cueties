```python
def get_word_length(df, list_of_tuples, s_c): 
    """
    Get word length for the content or source in the tuple (specified by s or c)
    """
    assert s_c == 's' or s_c == 'c'
    
    for t in list_of_tuples:
        s_tuple, c_tuple = t
        if s_c == 'c':
            start, end = c_tuple
            df.loc[start:end, 'c_length'] = (end+1-start)
        if s_c == 's':
            start,end = s_tuple
            df.loc[start:end, 's_length'] = (end+1-start)
```

```python
get_word_length(df, list_of_tuples, 'c')
get_word_length(df, list_of_tuples, 's')
```

```python
def get_distance_c2sentstart(df, list_of_tuples): 
    for t in list_of_tuples: 
        s, c = t
        start, end = c
        
        # Get the distance from the start of the document to the content span.
        c_doc = df.iloc[start]['filename']
        index_doc_start = df[df.filename == c_doc].first_valid_index()
        
        for i in range(start,end+1):
            df.loc[i, 'distance_c2docstart'] = i - index_doc_start
            
        # Get the distance from the start of the sentence to the content span.
        file_df = df.loc[df["filename"] == c_doc] # Filter by filename to get the correct indices.
        
        c_sent = df.iloc[start]["sentence_number"]
        index_sent_start = file_df[df.sentence_number == c_sent].first_valid_index()
        
        for i in range(start,end+1):
            df.loc[i, 'distance_c2sentstart'] = i-index_sent_start
```

```python
# Check whether S is in C

def check_s_in_c(df, list_of_tuples):
    for t in list_of_tuples: 
        s, c = t 
        s_start, s_end = s
        c_start, c_end = c 
        s_span = list(df.iloc[s_start:s_end+1]["lemma"])
        c_span = list(df.iloc[c_start:c_end+1]["lemma"])
        result = all(elem in s_span for elem in c_span)
        
        for i in range(c_start, c_end):
            df.loc[i, 's_in_c'] = result
```
