```python
'''PROPOSAL FOR NEW METHOD
FIND BEGINING QUOTES-take the following 2 sentences or up unitl finding an end or new begining quotation or document boundary.
FIND END QUOTES-take the previous 2 sentences or up until finding a begining or new end quotation or document boundary
COMPARE THE SPANS-compare these collected spans and remove any duplication of information (e.g. quotes found from both the
begining and end of their spans)'''
```

```python
def add_quotation(df):
    
    df['quotation'] = 'O'
    
    qmark_b_indices = list()
    qmark_e_indices = list()
    
  
    qmark_b_indices += list(df.loc[df['lemma']== '``' ].index)
    qmark_e_indices += list(df.loc[df['lemma']== "''"].index)
    
    print(len(qmark_b_indices))
    print(len(qmark_e_indices))
    assert len(qmark_b_indices) == len(qmark_e_indices)
        
    zipped_indices = zip(qmark_b_indices, qmark_e_indices)
    b_e_indices = list(zipped_indices)

    for b,e in b_e_indices:
        span = range(b, e)
        df.loc[span,'quotation'] = 'I'
        df.loc[b, 'quotation'] = 'B'
        df.loc[e, 'quotation'] = 'E'
            
    return df
```
