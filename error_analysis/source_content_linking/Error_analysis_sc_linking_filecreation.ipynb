```python
import pandas as pd
pd.set_option('display.max_columns', None)
from csv import DictReader
```

```python
import glob
file_list = glob.glob('./../../../Data/model_output_split_files/converted/polnear_dev/*')
```

```python
file_list
```

```python
df_gold = pd.read_csv('./../../../Data/Basline_df_correct\\Stella_newest_df_polnear.csv', sep='\t', index_col=0)
```

```python
df_gold.columns
```

```python
df_gold = df_gold[['filename', 'token', 'POS','doc_token_number', 'sentence_number', 'sentence_token_number', 
              'ne_info', 'relevant_ne', 'cue_label', 'candidate_source_label', 'attribution']]

df_gold.rename(columns = {'attribution': 'gold_attribution'}, inplace=True)
```

```python
df_pred = pd.read_csv(file_list[0], sep='\t', 
                 names=['filename', 'sentence_number', 'doc_token_number', 'sentence_token_number', 'token_in_doc', 'token', 'lemma','pos', 'dep1', 'dep2', 'pred_attribution'],
                index_col=False)
for file in file_list[1:]:
    df_temp = pd.read_csv(file, sep='\t', 
                 names=['filename', 'sentence_number', 'doc_token_number', 'sentence_token_number', 'token_in_doc', 'token', 'lemma','pos', 'dep1', 'dep2', 'pred_attribution'],
                index_col=False)
    df_pred = pd.concat([df_pred, df_temp])
    
#df_pred['filename'] = df_pred['filename'].apply(lambda row: row.replace('.xml', '.tsv'))
df_pred#.head()
```

```python
df_pred = df_pred[['filename', 'doc_token_number', 'pred_attribution']]
```

```python
df_pred.reset_index(drop=True, inplace=True)
```
