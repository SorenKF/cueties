```python
def quotation(df):
    '''fixing the quotation column'''
    
    df['quotation'] = 'O'
  
    qm_indices = list()

    qm_indices += list(df.loc[df['lemma'] == '``'].index)

    b_indices = list()
    e_indices = list()

    enumerated = enumerate(qm_indices)

    for index, item in enumerated:
        if index % 2 == 0:
            b_indices.append(index)
        else:
            e_indices.append(index)

    #if len(b_indices) == len(e_indices):
     #   continue
    #else:
        #print('ERROR, UNEVEN NUMBER OF QUOTATION MARKS WHEN SPLITTING')

    zipped_indices = zip(b_indices, e_indices)
    b_e_indices = list(zipped_indices)

    for b,e in b_e_indices:
        span = range(b, e)
        df.loc[span,'quotation'] = 'I'
        df.loc[b, 'quotation'] = 'B'
        df.loc[e, 'quotation'] = 'E'
            
    return df
```
