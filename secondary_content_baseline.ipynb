```python
'''function to find indirect quotation for the baseline system for content detection. The method of the function is to
    (1) find sentences which HAVE a cue but NOT direct quotation 
    (2) Split those sentences on the cue to find the before and 
    after spans 
    (3) take the longest portion of that sentence (either 
    before or after the cue) as the content using IOBE labels as with the direct quotation'''
```

```python
def candidate_indices(df):
    
    candidate_index_list= list()
    
    factors = list(zip(df['qm_in_sent'], df['cue_in_sentence']))
    enumerated = enumerate(factors)
    
    for index, pair in enumerated:
        
        qm_sent = pair[0]
        cue_sent = pair[1]
            
        if qm_sent == 1:
            continue
        elif cue_sent == 0:
            continue
        else:
            candidate_index_list.append(index)
        #print(len(candidate_index_list))
        '''candidate_index_list is a list containing the indices of tokens which have a cue but not direct quote in the sent'''
    return candidate_index_list
```

```python
candidate_index_list = candidate_indices(df)
```

```python
def sent_indices(df, candidate_index_list):
    
    section = df.loc[candidate_index_list]
    
    grouping_dict = defaultdict(list)

    for filename, sent_num, index in zip(section["filename"], section["sentence_number"],section.index):
        grouping_dict[filename, sent_num].append(index)
    return grouping_dict, section
```
